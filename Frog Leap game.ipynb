# **Frog Leap Puzzle Game**

**Overview:**


---


This script implements a text-based version of the Frog Leap puzzle game. The game involves moving frogs across a set of positions to solve the puzzle. The left set of frogs can only move right, and the right set of frogs can only move left. Frogs can move forward one space or jump two spaces by leaping over another frog from the opposite side. The puzzle is considered solved when the two sets of frogs have switched positions.

**Functions**:

---


1. `display_game(positions)`: Displays the current state of the Frog Leap puzzle.
    Parameters:
        - `positions` (list): A list representing the positions of frogs and an empty leaf.

2. `frog_leap_game()`: Implements the Frog Leap puzzle game logic.
    Game Flow:
        - Initializes the initial state of the puzzle with green ('G') and brown ('B') frogs.
        - Enters an infinite loop where the game state is displayed, and the player is prompted to enter a move.
        - Accepts user input and performs error checking.
        - Checks the validity of the selected move and updates the game state accordingly.
        - Checks for winning and losing conditions.
        - Breaks out of the loop if the player quits, reaches the end of the game, or wins.
    Input:
        - User enters the position of the frog to move (0-6) or 'q' to quit.
    Output:
        - Displays the current state of the game after each move.
        - Prints messages for invalid moves, reaching the end of the game, and winning.



  **How to Play the Game:**

1.Run the script to play the Frog Leap puzzle game.

2.Enter the position of the frog to move (0-6) or 'q' to quit.





````python
print("-----------------------------------------------FROG AND LEAP GAME--------------------------------------------------")

def display_game(positions):
    print("[  0 ,   1 ,   2 ,  3 ,  4 ,      5 ,   6 ]")
    print(positions)

def has_valid_moves(positions):
    for i in range(len(positions)):
        if positions[i] == '🟢':
            if i + 1 <= 6 and positions[i + 1] == '-':
                return True
            elif i + 2 <= 6 and positions[i + 2] == '-' and positions[i + 1] == '🟤':
                return True
        elif positions[i] == '🟤':
            if i - 1 >= 0 and positions[i - 1] == '-':
                return True
            elif i - 2 >= 0 and positions[i - 2] == '-' and positions[i - 1] == '🟢':
                return True
    return False

def frog_leap_game():
    while True:
        positions = ['🟢','🟢','🟢','-','🟤','🟤','🟤']

        while True:
            display_game(positions)

            move = input("Press 'q' to quit. Enter position of the frog to move: ")

            if move == 'q':
                print("You Lose")
                break

            try:
                move = int(move)
            except ValueError:
                print("Invalid input. Enter a valid position (0-6) or 'q' to quit.")
                continue

            if not 0 <= move <= 6:
                print("Invalid move. Enter a valid position (0-6).")
                continue

            if positions[move] == '-':
                print("Invalid move. Selected position has an empty leaf.")
                continue

            pos2 = 0

            if positions[move] == '🟢':
                if move + 1 <= 6 and positions[move + 1] == '-':
                    pos2 = move + 1
                elif move + 2 <= 6 and positions[move + 2] == '-' and positions[move + 1] == '🟤':
                    pos2 = move + 2
                else:
                    print("Invalid move.")
                    continue

            elif positions[move] == '🟤':
                if move - 1 >= 0 and positions[move - 1] == '-':
                    pos2 = move - 1
                elif move - 2 >= 0 and positions[move - 2] == '-' and positions[move - 1] == '🟢':
                    pos2 = move - 2
                else:
                    print("Invalid move.")
                    continue

            positions[move], positions[pos2] = positions[pos2], positions[move]

            # Check if empty leaf is at the end and no more valid moves
            if '-' in [positions[0], positions[6]] and not has_valid_moves(positions):
                print("Oops! You have reached the end of the game and no more valid moves.")
                reset_option = input("Do you want to reset the game? (y/n): ").lower()
                if reset_option == 'y':
                    print("-------------------------------------------------------------------------------------------------------")
                    print("                                           Resetting the game ")
                    print("-------------------------------------------------------------------------------------------------------")
                    break
                else:
                    print("You Lose")
                    break

            if positions == ['🟤', '🟤', '🟤', '-', '🟢', '🟢', '🟢']:
                display_game(positions)
                print("-------------------------------------------------------------------------------------------------------")
                print("                    🎉 Congratulations! You've triumphed in the Frog and Leap Game! 🏆")
                print("-------------------------------------------------------------------------------------------------------")
                break
        break
if __name__ == "__main__":
    frog_leap_game()
````
